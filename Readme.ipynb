# RateHub.ca : A Canadian company that helps Canadians make smarter educated financial decisions 

## Overview and Origin

* Name of company

Ratehub.ca


* When was the company incorporated?

The company is still privately held, the composition of capital is mainly venture capital and private investments. 


* Who are the founders of the company?

The company was founded by Alyssa Richard Furtado and James Laird August 1st 2010.

* How did the idea for the company (or project) come about?

Amidst the boom of online shopping, the company wanted to have offerings for the clients online to choose personal finance products with a simplified comparison, Ratehub started initially with credit cards and moved to have a larger range of personal finance products, ranging from mortgages, bank accounts, insurance for home and auto all the way to mortgages. 

* How is the company funded? How much funding have they received?

Initially The money went through two major funding rounds, a Seed funding round on August 2016 of 495 thousand dollars and a Series A funding round via a venture capital firm (Elephant VC) of $12M.

Worth noting that the company got strong offers from CBC's dragon's den but declined them and went the VC route.


## Business Activities:

* What specific financial problem is the company or project trying to solve?

When looking for a new personal finance product, you have to compare the products offered by several institutions and you have to either contact traditional financial agents in person or read long commercial presentations online; fir the uneducated person or the average citizen, there is many technicalities and fine prints that do not allow an informed decision when choosing a products.
Ratehub's mission is to help Canadians make smarter financial decisions.

* Who is the company's intended customer?  Is there any information about the market size of this set of customers?

The average Canadian citizen who is in the market for any personal finance product and looking to make the best choice/decision.
Per the latest company data available, they had over 6 million users in 2019.



* What solution does this company offer that their competitors do not or cannot offer? (What is the unfair advantage they utilize?)

The company is a double sided platform, they have the offerings for the clients and they go to the banks and financial institutions and get them to post their products and competitive rates on the website,
The second advantage the website has is that they offer their own insurance brokerage to the client as part of the options; and more recently they offer their in house mortgage lending through their subsidiary CanWise, a firm that managed to offer the lowest interest rate in the Canadian Mortgage history of 0.98%

* Which technologies are they currently using, and how are they implementing them? (This may take a little bit of sleuthing–– you may want to search the company’s engineering blog or use sites like Stackshare to find this information.)

The company uses several suites and technologies, the main platform for the Client resources management is Salesforce.

The company is focused on personal finance offering for the retail client, they do not offer solutions for companies or for institutional clients so far.

## Results

* What has been the business impact of this company so far?

More than 12 million Canadian are using the platform to make financial decisions and to obtain new products.
The company was growing successfully and was able to open their own Insurance Brokerage brand, their own Lending company (Canwise) and to acquire Moneysense (A very renown Canadian financial advice company existing since 1999) to put more emphasis on the financial advice part of the business.

* What are some of the core metrics that companies in this domain use to measure success? How is your company performing, based on these metrics?

The main metrics are the number of users and the annual revenue, in addition to the number of employees

	-The number of employees rose steadily over the years, currently the company employs 180 staff.
	-The company have an average of 1.2 million monthly users
	-The company latest reported annual revenue was 19.7 million dollars as per the publicly available data; alone, CanWise Financial, the company's mortgage broker funded almost 8 billion dollars in 2020.


* How is your company performing relative to competitors in the same domain?

Being the pioneer on the Canadian market and the only company that has their own brokerage from which they can offer very competitive rates, the company is household name when it comes to the offerings it has, there is several upcoming competitors but for now no one is matching their user base nor their broad offering of products.

Notable competitors are Rates.ca ; Homewise ; RateSupermarket.ca


## Recommendations

* If you were to advise the company, what products or services would you suggest they offer? (This could be something that a competitor offers, or use your imagination!)

Looking into offering investment opportunities in crypto currency and comparing the several Canadian Cryptocurrency exchanges would be a plus the company can offer.

Currently there is at least 7 accredited Canadian Cryptocurrency exchanges operating in Canada under Canadian regulations
 

* Why do you think that offering this product or service would benefit the company?

There is a large market for Canadians investing in cryptocurrencies, currently the users either go by word to mouth or have accounts with foreign international platforms operating overseas. With such a large untapped clientele, Ratehub can be the first to offer rational comparisons and draw clients and enthusiasts to Canadian exchanges.

* What technologies would this additional product or service utilize?

In addition to showing the rates for transfers and for buy/sell spread; Ratehub can have access to each crypto exchange public API to display a live comparison of the prices of the major Currencies at least side to side, that would be a major deciding factor to decide which exchange a client will settle with.

* Why are these technologies appropriate for your solution?

Public API are easily accessible and embeddable in any website and Ratehub could use the feature without disrupting their original layout, nor main display.

The Cryptocurrency exchange feature can be seamlessly added to the website.
